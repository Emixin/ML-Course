RBF Kernel Fixed X1 = 0

```python
import numpy as np
import matplotlib.pylab as plt


def rbf_kernel_fixed_x1_1d(x2, gamma):
    return np.exp(-gamma * (np.linalg.norm(x2)) ** 2)


x2_range = np.linspace(-5, 5, 100)
gammas = [0.5, 1, 5]
results = {}

for gamma in gammas:
    result = []
    for num in x2_range:
        result.append(rbf_kernel_fixed_x1_1d(num, gamma))
    results[gamma] = result

colors = ['red', 'blue', 'green']

plt.figure()
for gamma, color in zip(gammas, colors):
    plt.plot(x2_range, results[gamma], color=color)
plt.show()
```

RBF Kernel Fixed X1 = (0, 0)

```python
def rbf_kernel_fixed_x1_2d(x2, gamma=1):
    euclidean_dist = np.sum((x2 ** 2), axis=2)
    return np.exp(-gamma * euclidean_dist)


x_range = np.linspace(-3, 3, 100)
y_range = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_range, y_range)
x2 = np.stack((X, Y), axis=2)

result = rbf_kernel_fixed_x1_2d(x2)

plt.figure()
plt.contourf(X, Y, result)
plt.show()
```
